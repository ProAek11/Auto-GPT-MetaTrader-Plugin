# Framework

## Definitions, assumptions
* There is a tradable **target** time series $y(t)$ that we wish to learn and predict in order to trade and make profits. For example, in this project the target is Google stock.
* There are **features** saved in the design matrix $X(t)$, that we will use to predict the target. Feature building is discussed under  *Pre-processing* section.
* The prediction must be **causal**: the target on a time $t$, $y(t=t')$, can only be predicted by past information, $X(t<t')$! Notice the difference from non-temporal ML, where $X^{(i)}$ is used to learn/predict  $y^{(i)}$.
* The trading strategy must be **feasable in practice, in real time**.

## Algo-trading challenges
**Challenges characterizing sequential/time-series learning:**
* **Learning long term dependencies:** unlike modeling stationary, mean or simply frequent enough target dependancies on the inputs, it is hard to learn a hypothesis that "remembers" patterns for a long time since they occurred. From the book of Goodfellow et al., *Deep Learning*, 2016: "the problem of learning long-term dependencies remains one of the main challenges in deep learning".
* **Time dependent hypothesis**: even if the hypothesis captured well enough the target dependencies, unlike typical stationary learning tasks such as image recognition - it might be time dependent:
    * Different periods of the same length, even if long enough, can yield learnt hypotheses that substantially differ from one another.
    * The model validity might “expire” if a long time has passed since training.
    * Conclusions on the train/validation/test period lengths, in the context of learning or AI and decision making, are difficult to reach - changing or shifting periods might change the target dependencies found in them.
* **Time has direction and continuity that should be conserved in the learning**, for example splitting to train and validation/test sets cannot be done by random sampling of the data, train-validation periods should be continuous.
  
**Challenges characterizing algo-trading in particular:**
* **The bottom line is more important than the learning itself**. *The goal in algo-trading is not necessarily to have the smallest error in predicting the target*. For example, if a prediction that the target price will increase by 10% in the next time step led to opening a long, and the price went up "only" by 5%, then the prediction was "bad" (relative error of 100%) but the trading was profitable. An algorithm that misses the target, but in the right direction as in this example, can have consistently high error, and consistent profits.
* The goal in algo-trading is to derive a trading strategy which is **profitable** and **consistent** (**statistically significant**) on the test period(s).

## A heuristic for my basic (non-AI) trading strategy
My basic non-AI strategy - repeat each time a new prediction is made for the next time step:
* **go long (+1)** - if the target price is predicted to go up in a high enough confidence.
* **return to neutral (0)** - if the confidence for the prediction is not high enough.
* **go short (-1)** (if the target is shortable) - if the target price is predicted to go down in a high enough confidence.

Notes:
* **Confidence** - can be extracted from the learning algorithm, discussed below.
* Each position (long, neutral, short) is applied to all the existing portfolio, to have **exponential profits** (or losses).

---
Long & short in a nutshell:
* being long - having a positive amount of equity units, (for example, if the target is a stock - having a positive number of shares), profitable when the price increases relative to the price in which the long opened.
* being short - "the opposite of long", having a negative amount of equity units (=debt), profitable when the price decreases relative to the price in which the long opened. If needed, see an explanation about shorting in [Wikipedia](http://en.wikipedia.org/wiki/Short_(finance).


## Schemes for translating learning into trading predictions & decisions

Even though the target in algo-trading is continuous, the trading decisions in my basic trading strategy described above are discrete: go long, return to neutral, or go short.

My basic trading strategy does not care to know on each time step the next absolute target price, but the relative price difference from current price, and the prediction confidence level.

**I present here two schemes for using learning to predict and make trading decisions:**


### My simple regression trading scheme
Predicting the target price by regression (shallow or deep), translating to relative price movements in %, then **translating to discrete trading decisions** -

If the next predicted relative price movement (%) is:
  * higher than confidence level - **go long**,
  * else if lower than the negative confidence level - **go short**,
  * else - **return to neutral**.
  
**Confidence level** can be:
  * simply set to zero, meaning going long if next relative price movement (%) is positive, otherwise go short.
  * found as a hyper-parameter.
  * **inferred statistically**, as ~std of the training error, already known during test time, where $error(t)=predicted(t)-true(t)$.
  
    For example, if the training error std is $\sigma=2\%$ and the confidence level is set to $\sigma=2\%$, then going long can be chosen only if predicted price movement is higher than confidence level, $predicted(t)>\sigma=2\%$. Loss will occur if the true price movement is negative, $true(t)<0\%$, which will cause $TestError(t)=predicted(t)_{>\sigma}-true(t)_{<0\%}>\sigma$. Assuming the test error is similar to the train error which is approximately Gaussian, the probability of loss is therefore $P(TestError>\sigma)\simeq{P(TrainError>\sigma)}\approx{1/2(1-0.683)}\approx{15.9\%}$

### My simple classification trading scheme
Calculating relative price movements (%) from target prices, **discretizing the ground truth target** into +1,0,-1 classes representing the ground truth trading decisions, then predicting the classes which directly represent the trading decisions to be taken!
  
Discretizing the target - if the target relative price movement (%) is:
  * higher than confidence level: **up (go long) class (+1)**,
  * else if lower than the negative confidence level - **down (go short) class (-1)**,
  * else - **neutral class (0)**.
  
**Confidence level** can be:
  * simply set to zero, meaning positive relative price movement (%) is +1 class, otherwise -1 class.
  * found as a hyper-parameter.
  * **inferred statistically**, in a similar spirit as inferring the confidence level in regression: as ~std of the training % price movement distribution, already known during test time. Notice difference from the regression confidence (there it was ~std of the training *error*).

### My classes structure
I constructed ***regression* and *classification* classes** to perform all the calculations required to go from data to predictions and predicted trading decisions.

Both classes lack training methods, which are defined in:

### Model complexity considerations
Rich enough learning algorithms, as deep neural networds, should learn clever patterns even from raw data. However,** supplying complex features built out of raw data in a way that suits the learning task can be beneficial:**
* Improve the learning by making complex patterns more apparent, even for simple models.
* Shorten the optimization path that might have been required to learn the same complex features (that were supplied) directly from raw data.

**The possible downside of supplying complex features - over-fitting.**

As I already confessed, I am very methodical, and it's also my physicist's approach to start from the basics, and only if required add complexity (**"first simple, then correct"**). Therefore, **in this project we will use simple features, start from shallow learning and continue to more complex algorithms**. I am not interested to share all my ideas for feature building, but when it is relevant I will compare with my own results based on cleverer features.

### Causality (data leakage) considerations
**All features must maintain causality:** the target on a time $t'$, $y(t=t')$, can only be learnt/predicted by past information, $X(t<t')$!

However, the convention in ML is that each row of the design matrix $X$ contains the features to learn/predict the corresponding row of $y$. Therefore, this code will split the raw data in *data_df* into the the design matrix $X(t)$ and the target $y(t)$, shift the features one time step back relative to the target while keeping the index fixed, such that $X(t)=(x_0(t-\Delta t)),...,x_k(t-\Delta t)))$ can be used to learn/predict $y(t)$.

---
To verify that feature calculations don't violate causality, **I invented a simple verification: white noise regression**. There is a [sub-section](#scrollTo=IKJ4AvqLsoui) presenting it. In a nutshell: regression on completely random inputs taken from a distribution $N(0,\sigma^2)$ must yield a completely random prediction, and therefore error distributed as $N(0,\sigma^2)$ - or else cheating with information from the future (=breaking causality)!

### Building relative features

In building relative features from raw data, I mean that out of each raw input $x(t)$, **the relative feature that is built $x_{rel}(t)$ is the relative difference (in %) of $x(t)$ from its value on the previous time step: $x_{rel}(t)\equiv100\%\times\frac{x(t)-x(t-dt)}{x(t-dt)}$.**

Please notice that $x_{rel}(t)$ is **not** the (numeric) derivative of $x(t)$, which is $\frac{dx(t)}{dt}=\frac{x(t+dt)-x(t-dt)}{2dt}$

---
**Learning the relative features and the relative target can be quite useful in time series data when accumulated differences result in irrelevant bias, and only relative differences are important.** For example, if $x(t)$ is the price of a stock in certain period, then learning and trading according to $x(t)$ should be equivalent to using $x(t)+bias$, it is irrelevant if the stock started in $\$1$ or $\$100$ if it behaved the same in the whole period.

If needed, see [Histograms](#scrollTo=9gae15gBRRjj) section for a demonstration.

## Target temporal correlations
**In order to better understand the target temporal behavior and patterns, we will examine the target auto-correlation function *only in the train period*** - since it will give quantitative results that can be used in the learning, and should not be known in advance about the test period.

---
The target auto-correlation function $C(d)$ measures the temporal correlation between the target $y(t)$, and a copy of it shifted by $d$ time-steps, $y(t+d\times dt)$:

$C(d)\equiv corr(y(t),y(t+d\times dt))$, where $corr(a,b)\equiv\frac{\langle(a-\langle a\rangle)(b-\langle b\rangle)\rangle}{\sigma_a\sigma_b}\in [-1,1]$

Correlation of +1 between variables means they are completely correlated, they move the same (proportionally), and correlation of -1 means they are completely anti-correlated, they move in opposite directions.

**We can use the auto-correlation to find:**
* **Periodic patterns** - a peak in $C(d=period)$.
* **The auto-correlation decay length** - the typical length after which the auto-correlation decays from $C(d=0)=1$. This *might* be the number of past target time steps needed each time to predict future steps.

---
The target auto-correlation function in the train period:

````
max_corr_shift_ratio=0.75 # calculating C(d) when d goes from 0 to max_corr_shift_ratio*len(input)
# ------------------ end of inputs ------------------

corr=OzML.auto_correlation(train_test_split_obj.y_train_series.values,
                           max_corr_shift_ratio=max_corr_shift_ratio)
````

We can **extract the auto-correlation decay length, that can be defined as the length $\xi$ at which the correlation decays by half**, $C(d\equiv\xi)\equiv0.5C(0)=0.5$. From this graph we can extract $\xi\sim150$. This might be used later in the learning, as the number of past target time steps needed each time to predict future steps.

---
We observe a clear peak at a shift of ~1000 time steps, meaning that there is a periodic pattern in the target on train period, with period of ~1000 time steps. **Since 1000 days $\approx$ 4 trading years** (the mean annual trading days in NYSE and NASDAQ is ~250), **this periodicity is too long to be useful in this project** for daily trading.

### ElasticNet
ElasticNet is "nothing more" than a linear regression with L1-norm and L2-norm regularizations, however:
* I found the geometric interpretation of norm regularizations interesting.
* It can be used for **features post-selection** - by forcing a sparse hypothesis with minimal number of non-zero coefficients, it reveals what are the crucial features.

This comes with the expense of **adding two more free hyper-parameters to the learning** - the strengths of the two regularizations.

---
The optimization problem that ElasticNet solves:

$min(1/2n_{samples}\times||y - Xw||^2_2+\alpha r||w||_1+1/2 \times\alpha(1-r||w||^2_2)$

* $0<\alpha$ ('alpha' parameter) is the regularization strength ($\alpha=0$ means no regularization - linear regression, but in practice it is not recommended).
* $0\leq r\leq 1$ ('l1_ratio') is the ratio between L1 and L2 regularizations:
  * $r=1$ is a complete L1 regularization (Lasso),
  * $r=0$ is a complete L2 regularization (Ridge).

---
**Just to get a feeling about this algorithm we will use convenient regularization parameters I chose in advance. Of course that the proper way to find hyper-parameters is by cross validation** (see [Hyper-searching learning/trading parameters](#scrollTo=qe0a4pwoy-c9)).

````
EN_regression_obj=OzML.regression_shallow(train_test_split_obj)
EN_regression_obj.train_skElasticNet(l1_ratio=0.7,alpha=0.1)
````

First, we can compare the weights learnt now, with those learnt in linear regression:

We observe how the ElasticNet regularization forced a sparse hypothesis with 4 significant coefficients, compared to ~18 in the (un-regularized) [linear regression](#scrollTo=VGfTenevwLtL).

In my code, *regression_shallow* and *regression_deep* classes both inherit *regression* class, and similarly *classification_shallow* and  *classification_deep* inherit *classification*. **I constructed my classes such that it is equivalent to *use* both shallow and deep learning for making predictions and trading decisions**.

## Neural networks for regression
As I explained above, *using* deep learning with my code is not different than using shallow regression, for example [Linear regression](#scrollTo=VGfTenevwLtL). The difference is in the *learning*.

---
In order to train a neural network with my code, the next two lines will suffice:


```
regression_obj=OzML.regression_deep(train_test_split_obj)
regression_obj.train_TF_NN()
```

However, there are many hyper-parameters involved in training a neural network, so we will go through them:
* Gradient update will be done in batches:
  * ** *batch_size* ** is the number of data samples per gradient update.
  * ** *learning_rate* ** is the learning rate, should be tuned to have a well converging loss.
* I defined the training such that a training **epoch** is completing batch updates on all training samples. The maximal number of epochs until stopping training is ** *max_epochs* **.
* I defined a **callback** that stops the training when the validation loss has stopped improving on a validation period:
  * ** *validation_split* ** is the fraction of the training data to be used as validation data at the end of each epoch, selected from the end of the data.
  * ** *min_dloss_to_stop* ** is the minimum absolute change in the validation loss to qualify as an improvement. Since the loss is MSE (+regularization), the units are the target units squrated, $\$^2$.
  * ** *training_patience* ** is the number of epochs with no improvement after which training will be stopped
* The **optimizer** can be defined externally, else GradientDescentOptimizer defined in *regression_deep.train_TF_NN()* will be used.
* The **neural network model architecture** can be defined externally, else a default architecture defined in *regression_deep.train_TF_NN()* will be used.

### Shallow neural network - a sanity check

Before diving in, I believe it is appropriate to start with a **sanity check: training a shallow regression neural network (1 layer, 1 neuron), which is basically a linear regression – and verifying it reproduces the results we got in [Linear regression](#scrollTo=VGfTenevwLtL)** (otherwise – there is a problem!).

Therefore, I chose the neural network parameters to match linear regression:
* No validation period.
* Gradient descent optimizer.



````
# learning parameters
batch_size=round(0.1*len(train_test_split_obj.y_train_series)) # number of samples per gradient update
learning_rate=1e-2
max_epochs=1000

# callback parameters - stops training when a monitored quantity (val_loss) has stopped improving
validation_split=0 # fraction of the training data to be used as validation data at the end of each epoch, selected from the end of the data
min_dloss_to_stop=0.01**2 # minimum abs change in the monitored quantity to qualify as an improvement
training_patience=10 # number of epochs with no improvement after which training will be stopped

# optional- building the optimizer externally
import tensorflow as tf
optimizer=tf.train.GradientDescentOptimizer(learning_rate=learning_rate)

# optional - building the NN model externally
from tensorflow import keras

model=keras.Sequential()
model.add(keras.layers.Dense(1))

# optional - building the callback externally
callback=keras.callbacks.EarlyStopping(monitor='loss',
                    patience=training_patience,min_delta=min_dloss_to_stop)

# ------------------ end of inputs ------------------

shallow_NN_regression_obj=OzML.regression_deep(train_test_split_obj)
shallow_NN_regression_obj.train_TF_NN(validation_split,model,optimizer,callback=None,
    max_epochs=max_epochs,batch_size=batch_size,verbose=0)
shallow_NN_regression_obj.plot_training()
````

The neural network training seems to work well, the loss decreased smoothly.

* In [Linear regression](#scrollTo=VGfTenevwLtL) we got $\sqrt{MSE}\approx\$4.8$.
* The shallow neural network reached $\sqrt{MSE}\approx\$4.9$ with the current training parameters.

This seems close enough, we do not care to wait more time (to complete more epochs) for this demonstration to get a closer result.

Let's have a look at the prediction of this shallow neural network, to compare with the linear regression prediction, and mostly as a reminder of the baseline for the rest of the section:

## LSTM


### Short intro

I learnt this subject from the book of Goodfellow et al., *Deep Learning*, 2016, which was the best source I found for this subject – quite well organized, explains quite well. As I quoted from it in the section of [Algo-trading challenges](#scrollTo=4pQMpWfs47DX): **"the problem of learning long-term dependencies remains one of the main challenges in deep learning"**.

**Recurrent neural networks (RNNs)** are neural networks meant for learning *sequential* data, in which each state of target and inputs *generally* depends on the previous state in the sequence. The recurrent part in the network is the **self loop** $f$ that learns to relate each state to the previous one, $h^{(t)}=f(h^{(t-1)},X^{(t)};\theta)$

While RNNs are generally better suited to treat sequential data than ordinary NNs, **the challenge of long-term dependencies arises in RNNs because gradients propagated over many states tend to vanish **(or explode, rarely but with much damage).

---
To understand the problem of decaying or exploding gradients in RNNs you need to be familiar with linear algebra (otherwise feel free to skip this paragraph): neglecting inputs and activations, the self loop in a RNN repeats the same linear transform $W$ many times: $h^{(t)} =W^T h^{(t-1)}=(W^T )^t h^{(0)}$. Diagonalizing the transform $W=QΛQ^T$ gives $W^n=(QΛQ^T)^n=QΛ^n Q^T$, meaning that *any component of the transformation not aligned with the eigenvector with the largest eigenvalue will eventually decay relative to it*.

---
**Gated RNNs** are generally meant to treat the challenge of learning long-term dependencies by adding gates that control the self loop learning, and learn to remember and forget learnt dependencies.

**LSTM (Long Short Term Memory)** is a widely used and celebrated architecture of gated RNNs, in which the self loop is a leaky unit, a linear transform relates to the previous state, and gated.


### Training

Building RNNs in general, and an LSTM network in particular, is more complicated than just adding layers in TensorFlow as demonstrated previously for NN (and dropout) layers.

There is a fundamental difference from ordinary NNs, including ConvNets: when each step of the calculation depends on the previous one, as is the case in general RNN architectures and LSTM in particular, **training and prediction must be calculated serially (step-wise, batch_size=1) and cannot be parallelized**.

---
I will demonstrate LSTM for regression. As previously, complexity should be added gradually - so we can start with a simple and quite arbitrary architecture and minimal number of epochs - to avoid long training:

````
# learning parameters
max_epochs=10
optimizer=tf.train.GradientDescentOptimizer(learning_rate=1e-5)
validation_split=0.3 # fraction of the training data to be used as validation data at the end of each epoch, selected from the end of the data

# optional - building the NN model externally
model=keras.Sequential()
batch_input_shape=(1,1,len(train_test_split_obj.X_train_df.columns)) # to enable statefulness, this is the input shape, (batch_size, timesteps, input_dim)
model.add(keras.layers.LSTM(units=10,
    batch_input_shape=batch_input_shape,stateful=True,
    activation='tanh',recurrent_activation='hard_sigmoid'))
model.add(keras.layers.Dense(units=50,activation='relu'))
model.add(keras.layers.Dense(units=50,activation='relu'))
model.add(keras.layers.Dense(units=1))

# optional - building the callback externally
callback=None
# ------------------ end of inputs ------------------

LSTM_regression_obj=OzML.regression_deep(train_test_split_obj)
LSTM_regression_obj.train_TF_LSTM(validation_split,model,optimizer,callback,
    max_epochs=max_epochs,verbose=0)
LSTM_regression_obj.plot_training()
````

As expected, with step-wise calculation the LSTM took longer to train than a simple NN - in the previous section training was done in rate of ~0.1 sec/epoch, here it was done in an order of magnitude higher, ~5 sec/epoch.

We are not interested in dealing with the over-fitting here, and there is not much further to say about this short training of 10 epochs - but it will serve us to demonstrate prediction making:

### Making static and online predictions
Making predictions is also different from ordinary NNs - in LSTMs prediction is done step-wise, and also requires preparation before test prediction (first the train period should be predicted, to build up the state of the network as the initial state for the test period).

For this reason I built .predict_dynamically() method in my *regression_deep* class (until now we have used the .predic() method inherited from *regression* class). I used this opportunity to incorporate in this method an optional **online learning** for the test period, for both ordinary NNs and LSTMs.

First, let's see the LSTM prediction with no online learning:

````
LSTM_regression_obj.predict_dynamically(LSTM=True,online_learning_in_test=False)
LSTM_regression_obj.plot_prediction(rolling_window=0);
````

We are not surprised to see the over-fitting, that starts from the validation period (last 30% of the train period, since we set *validation_split=0.3*).

Now we can compare with online learning for the test period  - on each test sample first predicting the target, saving to the prediction results, and only then using this sample to further train the NN. **Online learning done in this way maintains causality and can be used in practice.** Let's see:

````
LSTM_regression_obj.predict_dynamically(LSTM=True,online_learning_in_test=True)
LSTM_regression_obj.plot_prediction(rolling_window=0);
````

We observe that **online learning improved the prediction in the test period** (which still suffers from severe over-fitting, of course).

## Causality / data leakage verification by white noise regression

**Time series learning must maintain causality:** the target on a time $t'$, $y(t=t')$, can only be learnt/predicted by past information, $X(t<t')$.

To verify that data preperation and learning calculations do not violate causality (data  leakage), **I invented a simple verification - white noise regression**: regression on completely random inputs taken from a distribution $N(0,\sigma^2)$ must yield a completely random prediction, and therefore error distributed as $N(0,\sigma^2)$ - or else cheating with information from the future (=breaking causality)!

---
We will use linear regression for this purpose, since my *regression* class uses the regressor to make predictions in the same way for all learning algorithms - the only difference between different learinng algorithms is the training (which uses scikit learn / TensorFlow standard algorithms - we can assume they maintain causailty).

````
white_y_std=5
# ------------------ end of inputs ------------------

# creating white noise from imported data in this project (data_df)
(white_data_df,white_attribute_names_list,white_noise_target_name)=OzML.whiten(
    data_df,
    white_y_mean=50,white_y_std=white_y_std,white_X_mean=100,white_X_std=10,
    frequency='1H',seed=0)

# creating features from the white data
(white_X_df,white_y_series)=OzML.basic_feature_building(white_data_df,
    white_noise_target_name,
    white_attribute_names_list,include_target_history_as_features)

# splitting data, normalizing (by the same inputs we used in this project's splitting)
white_train_test_split_obj=OzML.train_test_split(white_noise_target_name,'% differences')
white_train_test_split_obj.split_by_dates(white_X_df,white_y_series,
    train_start_datetime=y_series.index[0],
    train_duration_years=6.5,test_duration_years=6.5)
white_train_test_split_obj.normalize()

# linear regression
white_regression_obj=OzML.regression_shallow(white_train_test_split_obj)
white_regression_obj.train_skLinearRegression()
white_regression_obj.predict()
ax1,ax2=white_regression_obj.plot_prediction(rolling_window=10000)
ax1.set_ylabel('$')
ax2.set_ylabel('$');
````

We set *white_y_std=5*, target white noise std of $\sigma=5$, and indeed got a completely random prediction and error, with error std of $\$4.986=0.997\sigma$ for the train, and $\$4.99=0.998\sigma$ for the test, as expected (the white data length is $N=111385\sim10^5$ so the relative error of the std should be $1/\sqrt{N}\approx0.3\%$). **My code maintains causality**.
